```python
import pickle
```

```python
dict_gopigo = pickle.load(open("gopigo_connectome/connectome_dict.pickle","rb"))
dict_dataset = pickle.load(open("CElegansNeuronTables/connectome_dict.pickle","rb"))
```

```python
divergent = {}
for cell_in in dict_gopigo:
    if cell_in not in dict_dataset:
        divergent[cell_in] = dict_gopigo[cell_in]
    else:
        for cell_out in dict_gopigo[cell_in] :
            if cell_out not in dict_dataset[cell_in]:
                divergent[cell_in] = {cell_out: dict_gopigo[cell_in][cell_out]}
            else:
                if dict_gopigo[cell_in][cell_out] != dict_dataset[cell_in][cell_out]:
                    divergent[cell_in] = {cell_out: [dict_gopigo[cell_in][cell_out], dict_dataset[cell_in][cell_out]]}
                
```
